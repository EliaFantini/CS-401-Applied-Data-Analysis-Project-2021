```python
import textstat as ts # pip install textstat
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load pickle
SPEAKER_WITH_AFFILIATION_PATH = "../data/binary/data-2015.pickle"
speaker_with_affiliation = pd.read_pickle(SPEAKER_WITH_AFFILIATION_PATH)
speaker_with_affiliation.head()
```

```python
#split into democrats and republicans
#Q29468 is the republican party, Q29552 is the Democratic party
democrats = speaker_with_affiliation[speaker_with_affiliation["party"] == "Q29552"]
republicans = speaker_with_affiliation[speaker_with_affiliation["party"] == "Q29468"]
```

```python
democrats= democrats[democrats['quoteID'].isin(democrats["quoteID"].value_counts()[:1000].index.tolist())]
democrats = democrats.drop_duplicates(subset='quoteID')

republicans= republicans[republicans['quoteID'].isin(republicans["quoteID"].value_counts()[:1000].index.tolist())]
republicans = republicans.drop_duplicates(subset='quoteID')
```

```python
#df_obj = df_obj.assign(Percentage=lambda x: (x['Marks'] / x['Total']) * 100)
republicans['flesch_reading_easee']=[ ts.flesch_reading_ease(x) for x in republicans['quotation'] ]
republicans['gunning_fog']=[ ts.gunning_fog(x) for x in republicans['quotation'] ]
republicans['automated_readability_index']=[ ts.automated_readability_index(x) for x in republicans['quotation'] ]
republicans['coleman_liau_index']=[ ts.coleman_liau_index(x) for x in republicans['quotation'] ]
republicans['linsear_write_formula']=[ ts.linsear_write_formula(x) for x in republicans['quotation'] ]
republicans['dale_chall_readability_score']=[ ts.dale_chall_readability_score(x) for x in republicans['quotation'] ]
republicans['dale_chall_readability_score_v2']=[ ts.dale_chall_readability_score_v2(x) for x in republicans['quotation'] ]
republicans['text_standard']=[ ts.text_standard(x, True) for x in republicans['quotation'] ]
republicans.describe()
```

```python
democrats['flesch_reading_easee']=[ ts.flesch_reading_ease(x) for x in democrats['quotation'] ]
democrats['gunning_fog']=[ ts.gunning_fog(x) for x in democrats['quotation'] ]
democrats['automated_readability_index']=[ ts.automated_readability_index(x) for x in democrats['quotation'] ]
democrats['coleman_liau_index']=[ ts.coleman_liau_index(x) for x in democrats['quotation'] ]
democrats['linsear_write_formula']=[ ts.linsear_write_formula(x) for x in democrats['quotation'] ]
democrats['dale_chall_readability_score']=[ ts.dale_chall_readability_score(x) for x in democrats['quotation'] ]
democrats['dale_chall_readability_score_v2']=[ ts.dale_chall_readability_score_v2(x) for x in democrats['quotation'] ]
democrats['text_standard']=[ ts.text_standard(x, True) for x in democrats['quotation'] ]
democrats.describe()
```
