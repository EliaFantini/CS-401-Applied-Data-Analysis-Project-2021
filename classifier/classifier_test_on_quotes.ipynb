```python
import pickle
import pandas as pd
from classifier import preprocess_sentences
```

```python
# load the data
SPEAKER_WITH_AFFILIATION_PATH = "../data/binary/data-2017.pickle"
speaker_with_affiliation = pd.read_pickle(SPEAKER_WITH_AFFILIATION_PATH)
speaker_with_affiliation = speaker_with_affiliation.sample(10000)
X_test = list(speaker_with_affiliation['quotation'])
with open('topic_classifier', 'rb') as training_model:
    model = pickle.load(training_model)
```

```python
X_test = preprocess_sentences(X_test)
y = model.predict(X_test)
```

```python
speaker_with_affiliation['topic'] = y
lines = open("data/manifestolabels.txt").readlines()
topic_map = dict(map(lambda x: (int(x[3:6]), x[8:-2]),lines))
speaker_with_affiliation['topic'] = speaker_with_affiliation['topic'].map(topic_map)
speaker_with_affiliation.head(50)
```
